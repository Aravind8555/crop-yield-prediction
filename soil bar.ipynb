```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv( r"E:\Project\datasets\telangana_soil_data_2015_2025.csv")
```

```python
df .head()
```

```python
print(df.head())
```

```python
plt.rcParams['figure.figsize'] = (12, 6)

# Bar graph for pH by district
plt.figure()
df.groupby('District')['pH'].mean().sort_values().plot(kind='bar', color='skyblue')
plt.title('Average Soil pH by District')
plt.ylabel('pH')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Bar graph for Organic Carbon
plt.figure()
df.groupby('District')['Organic Carbon (%)'].mean().sort_values().plot(kind='bar', color='green')
plt.title('Average Organic Carbon (%) by District')
plt.ylabel('Organic Carbon (%)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Bar graph for Nitrogen
plt.figure()
df.groupby('District')['Nitrogen (kg/ha)'].mean().sort_values().plot(kind='bar', color='orange')
plt.title('Average Nitrogen (kg/ha) by District')
plt.ylabel('Nitrogen (kg/ha)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Bar graph for Phosphorus
plt.figure()
df.groupby('District')['Phosphorus (kg/ha)'].mean().sort_values().plot(kind='bar', color='purple')
plt.title('Average Phosphorus (kg/ha) by District')
plt.ylabel('Phosphorus (kg/ha)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Bar graph for Potassium
plt.figure()
df.groupby('District')['Potassium (kg/ha)'].mean().sort_values().plot(kind='bar', color='red')
plt.title('Average Potassium (kg/ha) by District')
plt.ylabel('Potassium (kg/ha)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
