```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_min = pd.read_csv(r"E:\Project\datasets\Monthly Minimum Temperature.csv")
df_max = pd.read_csv(r"E:\Project\datasets\Monthly Maximum Temperature.csv")
```

```python
df_min = df_min.drop(columns=['Mandal.1'], errors='ignore')
df_max = df_max.drop(columns=['Mandal.1'], errors='ignore')
```

```python
date_cols_min = df_min.columns[3:]
date_cols_max = df_max.columns[3:]
```

```python
df_min_avg = df_min[date_cols_min].mean(axis=0)
df_max_avg = df_max[date_cols_max].mean(axis=0)
```

```python
df_min_avg.index = pd.to_datetime(df_min_avg.index, format='%b-%y')
df_max_avg.index = pd.to_datetime(df_max_avg.index, format='%b-%y')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_min_avg.index, df_min_avg.values, marker='o', label='Min Temp (°C)', color='blue')
plt.plot(df_max_avg.index, df_max_avg.values, marker='o', label='Max Temp (°C)', color='red')
```
