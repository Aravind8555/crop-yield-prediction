```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files = [
    r"E:\Project\datasets\weather\TS Weather data June 2023.csv",
    r"E:\Project\datasets\weather\TS Weather data July 2023.csv",
    r"E:\Project\datasets\weather\TS Weather data August 2023.csv",
    r"E:\Project\datasets\weather\TS Weather data September 2023.csv"
]
```

```python
df_list = []
for file in files:
    df = pd.read_csv(file)
    df_list.append(df)

weather_data = pd.concat(df_list, ignore_index=True)
```

```python
weather_data.columns = [
    "District", "Mandal", "Date", "Rain(mm)",
    "Min Temp", "Max Temp",
    "Min Humidity(%)", "Max Humidity(%)",
    "Min Wind Speed(kmph)", "Max Wind Speed(kmph)"
]
```

```python
weather_data["Date"] = pd.to_datetime(weather_data["Date"], errors='coerce')
```

```python
weather_data["Month"] = weather_data["Date"].dt.strftime("%b-%Y")
monthly_avg = weather_data.groupby("Month").mean(numeric_only=True)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 12))

# Rainfall
monthly_avg[["Rain(mm)"]].plot(kind="bar", ax=axes[0], color="blue", legend=True)
axes[0].set_title("Average Rainfall per Month")
axes[0].set_ylabel("Rainfall (mm)")

# Temperature
monthly_avg[["Min Temp", "Max Temp"]].plot(kind="bar", ax=axes[1], legend=True)
axes[1].set_title("Average Temperature per Month")
axes[1].set_ylabel("Temperature (°C)")

# Humidity
monthly_avg[["Min Humidity(%)", "Max Humidity(%)"]].plot(kind="bar", ax=axes[2], legend=True)
axes[2].set_title("Average Humidity per Month")
axes[2].set_ylabel("Humidity (%)")

plt.tight_layout()
plt.show()

# Separate plot for Wind Speed
plt.figure(figsize=(12, 5))
monthly_avg[["Min Wind Speed(kmph)", "Max Wind Speed(kmph)"]].plot(kind="bar", figsize=(12, 5))
plt.title("Average Wind Speed per Month")
plt.ylabel("Wind Speed (kmph)")
plt.xlabel("Month")
plt.tight_layout()
plt.show()
```
